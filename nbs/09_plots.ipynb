```python
#exports
def plot_func(f, start, stop, num, logscalex=False, logscaley=False, figsize=(15,3), xlimit=None, ylimit=None): 
    plt.figure(figsize=figsize)
    if logscalex: x = np.logspace(start,stop,num,endpoint=True)
    else:         x = np.linspace(start,stop,num,endpoint=True)
    
    y = [f(x1) for x1 in x]
    if logscalex: plt.xscale('log')
    if logscaley: plt.yscale('log')
    plt.plot(x,y)
    
    if xlimit is not None: plt.xlim(*xlimit)
    if ylimit is not None: plt.ylim(*ylimit)
```

```python
plot_func(lambda x: x**3, -10, 10, 100)
```

```python
plot_func(lambda x: 10**x, -10, 10, 100,figsize=(10,1))
plot_func(lambda x: 10**x, -10, 10, 100,figsize=(10,1),logscaley=True)
```
